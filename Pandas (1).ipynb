## Panda Series

```python
import pandas as pd
import numpy as np
```

```python
marks = [34,76,89,23,67]
s = pd.Series(marks)
print(s)
print(type(s))
```

```python
marks = [34,76,89,23,67]
s = pd.Series(marks,["a","b","c","d","e"])
print(s)
print(type(s))
```

```python
marks = [34,76,89,23,67]
s = pd.Series(data=marks,index=["a","b","c","d","e"])
print(s)
print(type(s))
```

```python
marks = [34,76,89,23,67]
s = pd.Series(["a","b","c","d","e"],[34,76,89,23,67])
print(s)
print(type(s))
```

```python
marks = [34,76,89,23,67]
s = pd.Series(index=["a","b","c","d","e"], data=[34,76,89,23,67])
print(s)
print(type(s))
```

# Merging,Joining and Concateing

```python
import pandas as pd
import numpy as np
```

```python
# Create DataFrame
student = pd.DataFrame({"Name":list("ABCDEF"),
 "Age":list(range(23,29)),
 "Roll_num":[100,200,300,400,1000,900,],
 "Subject":["Math","Math","Biology","Art","Commerece","Biology"]})
student
```

```python
region = pd.DataFrame({"Roll_num":[100,200,300,400,500,600,700,800],
 "city":["Indore","Bhopal"]*4})
region
```

```python
pd.merge(left=student,
         right=region,
         how="inner",
         on="Roll_num")
```

```python
pd.merge(left=student,
         right=region,
         how="left",
         on="Roll_num")
```

```python
pd.merge(left=student,
         right=region,
         how="right",
         on="Roll_num")
```

```python
pd.merge(left=student,
         right=region,
         how="outer",
         on="Roll_num")
```

```python
pd.merge(left=student,
         right=region,
         how="outer",
         on="Roll_num",
        indicator=True)
```

```python
df_users = pd.DataFrame({
 'Year': [2019,2019,2019,2019,2019,2019,2019,2019,2019,2019,2019,2019,2020,2020,2020],
 'Quarter': ['Q1','Q1','Q1','Q2','Q2','Q2','Q3','Q3','Q3','Q4','Q4','Q4','Q1','Q1','Q1'],
 'Month': ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','Jan','Feb','Mar'],
 'Users': [150,170,160,200,190,196,210,225,260,210,212,219,630,598,321]
})
df_users

```

```python
 combine= pd.merge(left=df_users,right=df_ad_partners,how="left",on=["Year","Quarter"])
combine.head(5)
```
